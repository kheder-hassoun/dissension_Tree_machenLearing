About my DataSet ‚ù§   
--
This dataset contains information about various health-related factors   
(such as age, blood pressure, and creatinine levels) for individuals.  
The input features include age, anemia status, ejection fraction, and more.   
The output is whether an individual experienced a death event during the study. ü©∫üíî üò•



--------------


columns : 
---
Age: Represents the age of the individuals in the dataset.   
Anaemia: Indicates whether an individual has anemia (0 for no, 1 for yes).     
Creatinine Phosphokinase: Measures the level of an enzyme in the blood.      
Diabetes: Indicates whether an individual has diabetes (0 for no, 1 for yes).   
Ejection Fraction: Refers to the percentage of blood pumped out of the heart during each heartbeat.     
High Blood Pressure: Indicates whether an individual has high blood pressure (0 for no, 1 for yes).     
Platelets: Represents the count of platelets in the blood.      
Serum Creatinine: Measures the level of creatinine in the blood.   
Serum Sodium: Represents the concentration of sodium in the blood.     
Sex: Indicates the gender of the individuals (1 for male, 0 for female).     
Smoking: Indicates whether an individual smokes (0 for no, 1 for yes).      
Time: Represents the follow-up period in days.   

output :   
‚ö† DEATH_EVENT: Indicates whether an individual experienced a death event during the study (1 for yes, 0 for no).

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python

df = pd.read_csv("C:\\Users\\kheder\\Desktop\\machin_laerning_ex\\Desision_Tree\\heart_failure_clinical_records_dataset.csv")
df.head(4)
```

Now the decision Tree üòÅüå≤
--

```python
#feature selection ü•∞
feature_cols=['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction', 'high_blood_pressure', 'platelets', 'serum_creatinine','serum_sodium','sex','smoking','time']
x = df[feature_cols] # input
y = df.DEATH_EVENT # output ÿßŸÑÿπŸÖŸàÿØ ÿßŸÑŸÑŸä ÿßÿ≥ŸÖŸà DEATH_EVENT
```

```python
#split data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=47)# üòÅ 
```

```python
#decision tree üå≤
clf = DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
firstacc = metrics.accuracy_score(y_test,y_pred)
```

```python
print(" accurecy is : ",firstacc , "  üôÑ")
```

```python
from sklearn.tree import export_graphviz
from io import StringIO  
from IPython.display import Image  
import pydotplus

dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('diabetes.png')
Image(graph.create_png())
```

```python
from sklearn.model_selection import GridSearchCV
# Initialize the Decision Tree Classifier with possible parameters for tuning
parameters = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

clf = GridSearchCV(DecisionTreeClassifier(criterion='entropy'), parameters, cv=5)
clf.fit(x_train, y_train)

# Best parameters found
print(f'Best parameters: {clf.best_params_}')
```

```python

# Decision tree with best parameters
clf = DecisionTreeClassifier(max_depth=10, min_samples_leaf=2, min_samples_split=2)
clf = clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print(" accurecy after setting the best parameters is : ",metrics.accuracy_score(y_test,y_pred) , "  ‚ú®  it  was  : ", firstacc )
```

Why ??? 
--
The improvement in accuracy from 0.8111 to 0.8333    
can be attributed to the fact that the hyperparameter tuning process helped the decision tree learn a better representation of the data. 

-----

more details 
--
max_depth=10: A deeper tree (up to 10 levels) may have captured more complex patterns in the data.   
min_samples_leaf=2 and min_samples_split=2: These parameters prevent overfitting b   
 reducing the creation of very specific decision rules based on few data points.

let's draw the new tree üå¥ 
--

```python
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('diabetes2.png')
Image(graph.create_png())
```
